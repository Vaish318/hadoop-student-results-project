```python
%%writefile mapper_subject.py
#!/usr/bin/env python3
import sys
import csv

reader = csv.DictReader(sys.stdin)
for row in reader:
    try:
        subject = row['subject'].strip()
        marks = int(row['marks'])
        print(f"{subject}\t{marks}")
    except:
        continue
```

```python
%%writefile reducer_subject.py
#!/usr/bin/env python3
import sys
current = None
count = 0
total = 0

for line in sys.stdin:
    line = line.strip()
    if not line:
        continue
    subject, marks = line.split("\t")
    marks = int(marks)

    if current is None:
        current = subject

    if subject != current:
        print(f"{current}\t{count}\t{total}\t{100}\t{total/count}")
        current = subject
        count = 0
        total = 0

    count += 1
    total += marks

if current is not None:
    print(f"{current}\t{count}\t{total}\t{100}\t{total/count}")
```

```python
%%writefile mapper_student_safe.py
#!/usr/bin/env python3
import sys, traceback

def safe_print(s):
    try:
        sys.stdout.write(s+"\n")
    except:
        pass

for line in sys.stdin:
    line = line.strip()
    if not line: continue
    if line.startswith("student_id"): continue
    parts = line.split(",")
    if len(parts) < 5: continue

    sid = parts[0].strip()
    name = parts[1].strip()

    try:
        marks = int(parts[4].strip())
    except:
        continue

    safe_print(f"{sid}|{name}\t{marks}")
```

```python
%%writefile reducer_student.py
#!/usr/bin/env python3
import sys

current = None
total = 0

for line in sys.stdin:
    line = line.strip()
    if not line:
        continue

    parts = line.split("\t", 1)
    if len(parts) != 2:
        continue
    
    key, val = parts
    try:
        marks = int(val)
    except:
        continue

    if current is None:
        current = key
        total = 0

    if key != current:
        sid, name = current.split("|", 1)
        print(f"{sid}\t{name}\t{total}")
        current = key
        total = 0

    total += marks

if current is not None:
    sid, name = current.split("|", 1)
    print(f"{sid}\t{name}\t{total}")
```

```python
import os

def find_streaming_jar(start_dirs=("/Users/vaishnavi/hadoop", "/usr", "/opt"), max_depth=4):
    matches = []
    for start in start_dirs:
        for root, dirs, files in os.walk(start):
            # limit depth to avoid scanning entire disk
            if root.count(os.sep) - start.count(os.sep) > max_depth:
                continue
            for f in files:
                if f.startswith("hadoop-streaming-") and f.endswith(".jar"):
                    matches.append(os.path.join(root, f))
    return matches

# Only run fallback if the simple glob found nothing
if not streaming_jars:
    print("No jar found with the simple pattern; running fallback search (might be slower)...")
    streaming_jars = find_streaming_jar()
print("streaming_jars after fallback:", streaming_jars)
```

```python
# Cell 1 â€” Highest scoring student per subject (mapper+reducer)
import os, subprocess, shlex, glob, pandas as pd, io, sys

print("### Job: Highest-scoring student per subject\n")

HADOOP_HOME = os.environ.get("HADOOP_HOME", "/Users/vaishnavi/hadoop")
STREAMING_JAR = None
cand = glob.glob(os.path.join(HADOOP_HOME, "share","hadoop","tools","lib","hadoop-streaming-*.jar"))
if cand:
    STREAMING_JAR = cand[0]

# mapper: emits "subject \t name|marks"
mapper = """#!/usr/bin/env python3
import sys, csv
r = csv.DictReader(sys.stdin)
for row in r:
    try:
        subject = row['subject'].strip()
        name = row['name'].strip()
        marks = int(row['marks'])
        print(f"{subject}\\t{name}|{marks}")
    except Exception:
        continue
"""
# reducer: keep max marks per subject
reducer = """#!/usr/bin/env python3
import sys
current = None
best_name = None
best_marks = -1
for line in sys.stdin:
    line = line.strip()
    if not line: continue
    parts = line.split("\\t",1)
    if len(parts)!=2: continue
    subject, val = parts
    try:
        name, marks = val.rsplit("|",1)
        marks = int(marks)
    except:
        continue
    if current is None:
        current = subject
        best_name = name
        best_marks = marks
    if subject != current:
        print(f"{current}\\t{best_name}\\t{best_marks}")
        current, best_name, best_marks = subject, name, marks
    else:
        if marks > best_marks:
            best_marks = marks
            best_name = name
if current is not None:
    print(f"{current}\\t{best_name}\\t{best_marks}")
"""

open("mapper_max_subject.py","w").write(mapper)
open("reducer_max_subject.py","w").write(reducer)
os.chmod("mapper_max_subject.py", 0o755)
os.chmod("reducer_max_subject.py", 0o755)

local_input = "student_marks.csv"
hdfs_input = f"/user/{os.environ.get('USER','vaishnavi')}/student_data/student_marks.csv"
hdfs_out = f"/user/{os.environ.get('USER','vaishnavi')}/output_max_subject"

def run_local():
    cmd = f"cat {local_input} | python3 mapper_max_subject.py | sort | python3 reducer_max_subject.py"
    p = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    return p.stdout

def run_hadoop():
    # remove output, run streaming
    subprocess.run(shlex.split(f"hdfs dfs -rm -r -f {hdfs_out}"), capture_output=True)
    cmd = f"hadoop jar {STREAMING_JAR} -files mapper_max_subject.py,reducer_max_subject.py -mapper mapper_max_subject.py -reducer reducer_max_subject.py -input {hdfs_input} -output {hdfs_out}"
    subprocess.run(shlex.split(cmd), check=True)
    out = subprocess.check_output(shlex.split(f"hdfs dfs -cat {hdfs_out}/part-*"))
    return out.decode()

try:
    if STREAMING_JAR:
        print("Using Hadoop streaming jar:", STREAMING_JAR)
        raw = run_hadoop()
    else:
        print("Hadoop streaming JAR not found â€” running locally as fallback.")
        raw = run_local()
except Exception as e:
    print("Hadoop run failed â€” falling back to local. Error:", e)
    raw = run_local()

# load into pandas
df = pd.read_csv(io.StringIO(raw), sep="\t", header=None, names=["subject","name","marks"])
display(df.sort_values(["subject"]))
```

```python
# Cell 3 â€” Subject difficulty ranking (avg, pass percentage, count)
import os, subprocess, shlex, glob, pandas as pd, io

print("### Job: Subject difficulty ranking (average marks, pass percentage, count)\n")

# mapper prints: subject \t marks
mapper = """#!/usr/bin/env python3
import sys, csv
r = csv.DictReader(sys.stdin)
for row in r:
    try:
        subject = row['subject'].strip()
        marks = int(row['marks'])
        print(f"{subject}\\t{marks}")
    except:
        continue
"""
# reducer computes count, total, pass_count, avg, pass%
reducer = """#!/usr/bin/env python3
import sys
current=None
count=0
total=0
pass_count=0
for line in sys.stdin:
    line=line.strip()
    if not line: continue
    parts=line.split("\\t",1)
    if len(parts)!=2: continue
    subj, val = parts
    try:
        marks = int(val)
    except:
        continue
    if current is None:
        current = subj; count=0; total=0; pass_count=0
    if subj != current:
        avg = total/count if count else 0
        pct_pass = (pass_count/count*100) if count else 0
        print(f"{current}\\t{count}\\t{total}\\t{avg:.2f}\\t{pct_pass:.2f}")
        current = subj; count=0; total=0; pass_count=0
    count += 1
    total += marks
    if marks >= 40:
        pass_count += 1
if current is not None:
    avg = total/count if count else 0
    pct_pass = (pass_count/count*100) if count else 0
    print(f"{current}\\t{count}\\t{total}\\t{avg:.2f}\\t{pct_pass:.2f}")
"""

open("mapper_subject_stats.py","w").write(mapper)
open("reducer_subject_stats.py","w").write(reducer)
os.chmod("mapper_subject_stats.py",0o755)
os.chmod("reducer_subject_stats.py",0o755)

HADOOP_HOME = os.environ.get("HADOOP_HOME","/Users/vaishnavi/hadoop")
streaming = next(iter(glob.glob(os.path.join(HADOOP_HOME,"share","hadoop","tools","lib","hadoop-streaming-*.jar"))), None)
local_in = "student_marks.csv"
hdfs_in = f"/user/{os.environ.get('USER','vaishnavi')}/student_data/student_marks.csv"
hdfs_out = f"/user/{os.environ.get('USER','vaishnavi')}/output_subject_stats2"

def run_local():
    cmd = f"cat {local_in} | python3 mapper_subject_stats.py | sort | python3 reducer_subject_stats.py"
    import subprocess
    p = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    return p.stdout

def run_hadoop():
    subprocess.run(shlex.split(f"hdfs dfs -rm -r -f {hdfs_out}"), capture_output=True)
    cmd = f"hadoop jar {streaming} -files mapper_subject_stats.py,reducer_subject_stats.py -mapper mapper_subject_stats.py -reducer reducer_subject_stats.py -input {hdfs_in} -output {hdfs_out}"
    subprocess.run(shlex.split(cmd), check=True)
    return subprocess.check_output(shlex.split(f"hdfs dfs -cat {hdfs_out}/part-*")).decode()

try:
    if streaming:
        print("Running on Hadoop streaming:", streaming)
        raw = run_hadoop()
    else:
        print("Hadoop streaming not found â€” running locally.")
        raw = run_local()
except Exception as e:
    print("Hadoop failed, local fallback. Err:", e)
    raw = run_local()

df = pd.read_csv(io.StringIO(raw), sep="\t", header=None, names=["subject","count","total_marks","avg_marks","pass_pct"])
df = df.sort_values("avg_marks")  # lower avg = harder
display(df)
```

```python
# Cell 5 â€” Student Performance Categories (Strong >75, Average 50-75, Weak <50)
import os, subprocess, shlex, glob, pandas as pd, io

print("### Job: Student performance categories (Strong/Average/Weak)\n")

# Mapper: emits student_id|name \t marks
mapper = """#!/usr/bin/env python3
import sys, csv
r = csv.DictReader(sys.stdin)
for row in r:
    try:
        sid = row['student_id'].strip()
        name = row['name'].strip()
        marks = int(row['marks'])
        print(f"{sid}|{name}\\t{marks}")
    except:
        continue
"""
# Reducer: sum marks per student, infer category using inferred max_total (if subjects unknown assume 7 subjects present; but we'll just use avg)
reducer = """#!/usr/bin/env python3
import sys
current=None
total=0
count=0
for line in sys.stdin:
    line=line.strip()
    if not line: continue
    parts=line.split("\\t",1)
    if len(parts)!=2: continue
    key,val=parts
    try:
        marks = int(val)
    except:
        continue
    if current is None:
        current = key; total=0; count=0
    if key != current:
        sid,name = current.split("|",1)
        avg = total/count if count else 0
        if avg > 75: cat='Strong'
        elif avg >= 50: cat='Average'
        else: cat='Weak'
        print(f"{sid}\\t{name}\\t{total}\\t{count}\\t{avg:.2f}\\t{cat}")
        current = key; total=0; count=0
    total += marks
    count += 1
if current is not None:
    sid,name = current.split("|",1)
    avg = total/count if count else 0
    if avg > 75: cat='Strong'
    elif avg >= 50: cat='Average'
    else: cat='Weak'
    print(f"{sid}\\t{name}\\t{total}\\t{count}\\t{avg:.2f}\\t{cat}")
"""

open("mapper_perf_cat.py","w").write(mapper)
open("reducer_perf_cat.py","w").write(reducer)
os.chmod("mapper_perf_cat.py",0o755)
os.chmod("reducer_perf_cat.py",0o755)

HADOOP_HOME=os.environ.get("HADOOP_HOME","/Users/vaishnavi/hadoop")
streaming = next(iter(glob.glob(os.path.join(HADOOP_HOME,"share","hadoop","tools","lib","hadoop-streaming-*.jar"))), None)
local_in="student_marks.csv"; hdfs_in=f"/user/{os.environ.get('USER','vaishnavi')}/student_data/student_marks.csv"
hdfs_out=f"/user/{os.environ.get('USER','vaishnavi')}/output_perf_cat"

def run_local():
    cmd=f"cat {local_in} | python3 mapper_perf_cat.py | sort | python3 reducer_perf_cat.py"
    import subprocess
    return subprocess.run(cmd, shell=True, capture_output=True, text=True).stdout

def run_hadoop():
    subprocess.run(shlex.split(f"hdfs dfs -rm -r -f {hdfs_out}"), capture_output=True)
    cmd=f"hadoop jar {streaming} -files mapper_perf_cat.py,reducer_perf_cat.py -mapper mapper_perf_cat.py -reducer reducer_perf_cat.py -input {hdfs_in} -output {hdfs_out}"
    subprocess.run(shlex.split(cmd), check=True)
    return subprocess.check_output(shlex.split(f"hdfs dfs -cat {hdfs_out}/part-*")).decode()

try:
    raw = run_hadoop() if streaming else run_local()
except Exception as e:
    print("Hadoop error:", e)
    raw = run_local()

df = pd.read_csv(io.StringIO(raw), sep="\t", header=None, names=["student_id","name","total_marks","subjects_count","avg_marks","category"])
display(df.head(50))
# Also a count per category
display(df['category'].value_counts().rename_axis('category').reset_index(name='count'))
```

```python
# Cell 6 â€” Pairwise student similarity (LIMITED sample)
# WARNING: This is O(n^2). We limit to first N students to keep runtime reasonable.
import os, subprocess, shlex, glob, pandas as pd, io, itertools, csv, math

print("### Job: Pairwise student similarity (limited sample, difference <5 marks per subject)\n")
N = 200  # change if you know your env can handle more

# We'll run locally here â€” easier to implement and demonstrate.
# Strategy:
# 1) build a dict mapping student -> {subject: marks}
# 2) for first N students compute pairwise similarity = number of subjects where abs(diff)<5
# 3) output top pairs

data = {}
with open("student_marks.csv","r") as f:
    r = csv.DictReader(f)
    for row in r:
        sid = row['student_id'].strip()
        subj = row['subject'].strip()
        try:
            marks = int(row['marks'])
        except:
            continue
        data.setdefault(sid, {})[subj] = marks

students = sorted(data.keys())[:N]
pairs = []
for a,b in itertools.combinations(students,2):
    a_map = data[a]
    b_map = data[b]
    common = set(a_map.keys()) & set(b_map.keys())
    if not common:
        continue
    similar_count = sum(1 for s in common if abs(a_map[s]-b_map[s]) < 5)
    # percentage similarity over common subjects
    pct = similar_count / len(common) * 100
    pairs.append((a,b,similar_count,len(common),pct))

pairs_sorted = sorted(pairs, key=lambda x:(-x[2],-x[4]))[:50]
df = pd.DataFrame(pairs_sorted, columns=["student_a","student_b","similar_count","common_subjects","similar_pct"])
display(df.head(50))
print(f"\n(Computed over first {N} students; increase N carefully â€” complexity O(N^2)).")
```

```python
# Cell 7 â€” Subject-wise standard deviation (shows spread)
import pandas as pd, csv, io
print("### Job: Subject-wise standard deviation (spread of marks)\n")

subj_vals = {}
with open("student_marks.csv","r") as f:
    r = csv.DictReader(f)
    for row in r:
        subj = row['subject'].strip()
        try:
            m = float(row['marks'])
        except:
            continue
        subj_vals.setdefault(subj, []).append(m)

rows = []
import math
for s,vals in subj_vals.items():
    n = len(vals)
    mean = sum(vals)/n if n else 0
    var = sum((x-mean)**2 for x in vals)/n if n else 0
    std = math.sqrt(var)
    rows.append((s, n, mean, std))

df = pd.DataFrame(rows, columns=["subject","count","mean_marks","stddev_marks"]).sort_values("stddev_marks", ascending=False)
display(df)
```

```python
# Cell 8 â€” Count number of subjects where each student scored >80 (to find consistent high performers)
import csv, pandas as pd
print("### Job: Number of subjects per student where marks > 80\n")

counts = {}
with open("student_marks.csv","r") as f:
    r = csv.DictReader(f)
    for row in r:
        sid = row['student_id'].strip()
        name = row['name'].strip()
        try:
            m = int(row['marks'])
        except:
            continue
        counts.setdefault((sid,name), 0)
        if m > 80:
            counts[(sid,name)] += 1

rows = [(sid,name,c) for (sid,name),c in counts.items()]
df = pd.DataFrame(rows, columns=["student_id","name","subjects_gt80"]).sort_values(["subjects_gt80","student_id"], ascending=[False,True])
display(df.head(50))
```

```python
# Setup: run once
import os, subprocess, shlex, io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Image, Markdown

plt.rcParams['figure.figsize'] = (10,5)
sns.set(style="whitegrid")

print("Available screenshots (for use in your report):")
print("/mnt/data/Screenshot 2025-11-21 at 18.55.49.png")
print("/mnt/data/Screenshot 2025-11-21 at 18.55.57.png")
print("\nTrying to fetch HDFS outputs. If HDFS isn't available, this will use local CSVs (student_marks.csv, student_totals_graded.csv).")

USER = os.environ.get("USER","vaishnavi")
# HDFS output dirs we used earlier
hdfs_totals = f"/user/{USER}/output_student_totals"
hdfs_totals_jupyter = f"/user/{USER}/output_student_totals_jupyter"
local_marks = "student_marks.csv"
local_totals_graded = "student_totals_graded.csv"

def read_hdfs_cat(path_pattern):
    try:
        cmd = f"hdfs dfs -cat {path_pattern}"
        raw = subprocess.check_output(shlex.split(cmd), stderr=subprocess.DEVNULL).decode()
        return raw
    except Exception as e:
        return None

# prefer Hadoop totals if present
raw = read_hdfs_cat(f"{hdfs_totals}/part-*") or read_hdfs_cat(f"{hdfs_totals_jupyter}/part-*")
if raw:
    # raw is tab-separated lines like: S00014\tPriya\t31
    df_totals = pd.read_csv(io.StringIO(raw), sep="\t", header=None, names=["student_id","name","total"])
    print("Loaded student totals from HDFS:", df_totals.shape)
else:
    if os.path.exists(local_totals_graded):
        df_totals = pd.read_csv(local_totals_graded)
        # if the CSV has percent/grade columns we use them; else keep as-is
        if {"student_id","name","total"}.issubset(df_totals.columns):
            print("Loaded local student_totals_graded.csv:", df_totals.shape)
        else:
            # sometimes student_totals_graded.csv contains 5 cols (id,name,total,pct,grade)
            if df_totals.shape[1] >= 3:
                df_totals.columns = ["student_id","name","total","percent","grade"][:df_totals.shape[1]]
            print("Loaded local student_totals_graded.csv (mapped):", df_totals.shape)
    elif os.path.exists(local_marks):
        # compute totals from student_marks.csv if no totals
        df_marks = pd.read_csv(local_marks)
        df_totals = df_marks.groupby(["student_id","name"], as_index=False)["marks"].sum().rename(columns={"marks":"total"})
        print("Built totals by grouping local student_marks.csv:", df_totals.shape)
    else:
        raise FileNotFoundError("No HDFS output found and no local CSV (student_marks.csv or student_totals_graded.csv). Please upload or run streaming job first.")

# ensure numeric
df_totals['total'] = pd.to_numeric(df_totals['total'], errors='coerce').fillna(0).astype(int)

# Also read full marks (per-subject) if possible
raw_marks = read_hdfs_cat(f"/user/{USER}/student_data/student_marks.csv") or None
if raw_marks:
    df_marks = pd.read_csv(io.StringIO(raw_marks))
    print("Loaded student_marks.csv from HDFS:", df_marks.shape)
elif os.path.exists(local_marks):
    df_marks = pd.read_csv(local_marks)
    print("Loaded local student_marks.csv:", df_marks.shape)
else:
    df_marks = None
    print("Per-subject marks file not found locally or in HDFS â€” some visualizations will be skipped.")

# infer number of subjects (useful for percent)
if 'total' in df_totals.columns:
    max_observed = df_totals['total'].max()
    num_subjects = max(1, (max_observed + 99)//100)
    MAX_TOTAL = num_subjects * 100
else:
    MAX_TOTAL = 100
    num_subjects = 1
print(f"Inferred subjects={num_subjects}, max_total={MAX_TOTAL}")
```

```python
# Donut chart of grade distribution
print("### Grade distribution (Donut)")

# If df_totals has a grade field use it, else compute grade from percent
if 'grade' not in df_totals.columns:
    df_totals['percent'] = df_totals['total'] / float(MAX_TOTAL) * 100
    def grade(p):
        if p >= 90: return 'A'
        if p >= 75: return 'B'
        if p >= 60: return 'C'
        if p >= 40: return 'D'
        return 'F'
    df_totals['grade'] = df_totals['percent'].apply(grade)

grade_counts = df_totals['grade'].value_counts().reindex(['A','B','C','D','F']).fillna(0).astype(int)
fig, ax = plt.subplots(figsize=(6,6))
wedges, texts = ax.pie(grade_counts.values, labels=grade_counts.index, startangle=90, wedgeprops=dict(width=0.4))
ax.set_title("Grade distribution (A/B/C/D/F)")
# draw center circle to create donut
centre_circle = plt.Circle((0,0),0.4,fc='white')
ax.add_artist(centre_circle)
plt.tight_layout()
plt.show()
# save
fig.savefig("grade_distribution_donut.png")
print("Saved: grade_distribution_donut.png")
```

```python
# Pareto chart: top students cumulative contribution
print("### Pareto chart â€” cumulative contribution of top students to total marks")

df_sorted = df_totals.sort_values("total", ascending=False).reset_index(drop=True)
df_sorted["cum_sum"] = df_sorted["total"].cumsum()
df_sorted["cum_pct"] = df_sorted["cum_sum"] / df_sorted["total"].sum() * 100
df_sorted["rank"] = df_sorted.index + 1

# choose top N for plotting to keep legible (e.g., top 200 or all if small)
N = min(200, len(df_sorted))
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.bar(df_sorted["rank"].iloc[:N], df_sorted["total"].iloc[:N])
ax1.set_xlabel("Student rank")
ax1.set_ylabel("Total marks (bar)")

ax2 = ax1.twinx()
ax2.plot(df_sorted["rank"].iloc[:N], df_sorted["cum_pct"].iloc[:N], color='k', marker='o')
ax2.set_ylabel("Cumulative % of total marks")
ax1.set_title(f"Pareto: top {N} students contribution (total students = {len(df_sorted)})")
plt.tight_layout()
fig.savefig("pareto_top_students.png")
plt.show()
print("Saved: pareto_top_students.png")
```
